```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/Users/Claibourne/Downloads/cities_air_quality_water_pollution.18-10-2021 (1).csv")
```

```python
df.columns = ["City", "Region", "Country", "AirQuality", "WaterPollution"]
```

```python
plt.plot(df["WaterPollution"], "o-", 
         color="tab:blue", markersize=0.00000000000000000000000000000000000000000004)
```

```python

us_data = df[df['Country'].str.contains('United States of America', case=False, na=False)]
print(us_data)
```

```python
air = df["AirQuality"]
water = df["WaterPollution"]
plt.figure(figsize=(10, 6))
plt.plot(us_data["AirQuality"], "o-", color="tab:blue", markersize=4, label="Air Quality")
plt.plot(us_data["WaterPollution"], "--", color="tab:orange", alpha=0.7, label="Water Pollution")
plt.title('Air Quality vs Water Pollution in the ')
plt.xlim(min(air) - 1, max(air) + 1)
plt.ylim(min(water) - 1, max(water) + 1)
plt.grid(True)
plt.show()





```
